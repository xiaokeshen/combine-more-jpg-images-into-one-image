```python
"""
This program is used to combine the photoes in a specified folder into one jpg file. It is written by using python3.5
reference: 
[1]http://mibix.de/blog/photography/item/22-combine-several-images-into-one-with-python-and-pil

[2]http://stackoverflow.com/questions/29234413/too-many-open-files-error-when-opening-and-loading-images-in-pillow
"""

import os
from PIL import Image
#import Image
from PIL import ImageFont
from PIL import ImageDraw


img_list   = []
img_path   = "/Users/jimmy/Dropbox/data_viz/test_new/"
#width      = 647 #684
#height     = 402 #1024


width      = 588 #684
height     = 394 #1024

text_pos_x = 3   #10
text_pos_y = 3   #10
text_size  = 10  #50
rows       = 30
columns    = 0


for file in os.listdir( img_path ):
    # Search images
    if file.endswith(".jpg"):
        # open images
        temp = Image.open(img_path + file)
        keep = temp.copy()
        img_list.append(keep )
        temp.close()

for image in img_list:
    # resize images
    image.thumbnail( (width, height) )

columns =int(len(img_list)/rows)

#creates a new empty image, RGB mode
mosaic = Image.new( 'RGB', (  columns * width, rows * height ) )

# Prepare draw and font objects to render text
#draw   = ImageDraw.Draw(mosaic)
#font   = ImageFont.truetype("C:/Windows/Fonts/arial.ttf",text_size)

k=0
for j in range( 0, rows * height, height ):
    for i in range( 0, columns * width, width ):
        # paste the image at location i,j:
        mosaic.paste( img_list[k], (i,j) )
        # render text at location i,j:
        #draw.text((i + text_pos_x, j + text_pos_y), img_text[k], font=font)
        # Select next image and text
        k = k + 1

# Save image to file
mosaic.save('/Users/jimmy/Dropbox/data_viz/test_output/mosaic_25_31_v3.jpg')
```
